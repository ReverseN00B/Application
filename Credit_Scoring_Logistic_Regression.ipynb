Split Train & test

```python
# create Train Test Split
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split
# X = df.drop('TARGET', axis=1)
# y = df.iloc[:, 0].values
y = df.iloc[:, 0].values
X = df.iloc[:, 1:29].values

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.20,
                                                    random_state=0,
                                                    stratify = y)

# Applying Standardization Std Scaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

```

```python
# Exporting Normalisation Coefficients for later use in prediction
import joblib
joblib.dump(sc, '/content/Normalisation_CreditScoring')
```

Model Building

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_train, y_train)

# model.coef_
y_pred = model.predict(X_test)

# Exporting Logistic Regression Classifier for later use in prediction

# import joblib
joblib.dump(model, '/content/f1_Classifier_CreditScoring')
```

Model Evaluation


```python
from sklearn.metrics import confusion_matrix,accuracy_score

print(confusion_matrix(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```

Writing output file


```python
predictions = model.predict_proba(X_test)
predictions
```

```python
# writing model output file

df_prediction_prob = pd.DataFrame(predictions, columns = ['prob_0', 'prob_1'])
df_prediction_target = pd.DataFrame(model.predict(X_test), columns = ['predicted_TARGET'])
df_test_dataset = pd.DataFrame(y_test,columns= ['Actual Outcome'])

dfx=pd.concat([df_test_dataset, df_prediction_prob, df_prediction_target], axis=1)

dfx.to_csv("/content/Model_Prediction.csv", sep=',', encoding='UTF-8')

dfx.head()
```
